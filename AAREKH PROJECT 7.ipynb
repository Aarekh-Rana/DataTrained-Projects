```python
import numpy as np
for col in ds_baseball.columns:
    if ds_baseball.skew().loc[col]>0.55:
        ds_baseball[col]=np.log1p(ds_baseball[col])
```

```python
sns.pairplot(ds_baseball)
```

```python
from scipy.stats import zscore
zscore=abs(zscore(ds_baseball))
print(ds_baseball.shape)
ds_baseball1=ds_baseball.loc[(zscore<3).all(axis=1)]
print(ds_baseball1.shape)
```

```python
x=ds_baseball1
```

```python
y=ds_baseball1["W"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(30,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=r_state)
    lr=linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2_score corresponding to random state:",r_state,"is",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2_score corresponding to",final_r_state,"is",max_r_score)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring="r2").mean()
```
